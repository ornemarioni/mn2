### Distribucion homogenea en una esfera de radio 1

```python
import random
import numpy as np

N = 4096
random.seed(27)
sample = frange(0,1,1e-5)
    
r3 = random.sample(sample, N)
r3 = np.asarray(r3)

cos_t = random.sample(sample, N)
cos_t = np.asarray(cos_t) * 2. - 1

phi = random.sample(sample, N)
phi = np.asarray(phi) * 2. * np.pi 

sin_t = np.sqrt(1 - cos_t**2)

x = r3**(1/3.) * sin_t * np.cos(phi)
y = r3**(1/3.) * sin_t * np.sin(phi)
z = r3**(1/3.) * cos_t
```

## Kernel con soporte compacto (Ec. 4.21)

```python
import numpy as np

#Spline kernel of Monaghan and Lattanzio (1985)
def W(r,h):
    
    n=len(r)
    
    x = r/h
    W = np.zeros(n)
    
    for i in range(n):
        
        if 0.<= x[i] and x[i] < 1. :
            W[i] = (1. - 1.5*(x[i]**2) + 0.75*(x[i]**3)) / (np.pi * h**3)
        
        elif 1.<= x[i] and x[i] < 2. :
            W[i] = (2. - x[i])**3 / (4 * np.pi * h**3)
        
        else:
            W[i] = 0.
    
    return W
```

```python
h = 0.5
r = frange(0,2.,0.001)

import matplotlib.pyplot as plt
import numpy as np
from __future__ import unicode_literals
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.unicode'] = True

fig = plt.figure(1, figsize=(8,8))
fig.subplots_adjust(bottom=0.15, left =0.16, right = 0.95, top = 0.95)

ax1 = fig.add_subplot(111)
ax1.plot(r, W(r,h), '.k', markersize=5)
ax1.minorticks_on()
ax1.tick_params( labelsize=24)
ax1.tick_params('both', length=3, width=1.2,which='minor', direction='in', right='on',top='on')
ax1.tick_params('both', length=6, width=1.2,which='major', direction='in', right='on',top='on') 
ax1.set_xlabel('$\mathrm{r}$', fontsize=24)
ax1.set_ylabel(r'$\mathrm{W(r,h)}$', fontsize=24)


plt.show()
```

#### Gradiente del kernel W(r,h)

```python
def dW(r,h):

    n=len(r)
    x = r/h
    dW = np.zeros(n)
    
    for i in range(n):
        
        if 0.<= x[i] and x[i] < 1. :
            dW[i] = 3 * ((0.75 * x[i]**2) - x[i]) / (np.pi * h**4)
        
        elif 1.<= x[i] and x[i] < 2. :
            dW[i] = -3. * (2. - x[i])**2 / (4 * np.pi * h**4)
        
        else:
            dW[i] = 0.
    
    return dW
```

```python
h = 0.5
r = frange(0,2.,0.001)

import matplotlib.pyplot as plt
import numpy as np
from __future__ import unicode_literals
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.unicode'] = True

fig = plt.figure(1, figsize=(8,8))
fig.subplots_adjust(bottom=0.15, left =0.16, right = 0.95, top = 0.95)

ax1 = fig.add_subplot(111)
ax1.plot(r, dW(r,h), '.k', markersize=5)
ax1.minorticks_on()
ax1.tick_params( labelsize=24)
ax1.tick_params('both', length=3, width=1.2,which='minor', direction='in', right='on',top='on')
ax1.tick_params('both', length=6, width=1.2,which='major', direction='in', right='on',top='on') 
ax1.set_xlabel('$\mathrm{r}$', fontsize=24)
ax1.set_ylabel(r'$\mathrm{\nabla W(r,h)}$', fontsize=24)


plt.show()
```
