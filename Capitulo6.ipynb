# Con viscosidad y sin viscosidad

```python
data   = np.loadtxt(path+'lh1.out')
r   = data[:,2]
vel = data[:,3]
rho = data[:,4]
P   = data[:,5]
e   = data[:,6]
Q   = data[:,7]
```

```python
fig=plt.figure(1, figsize=(8,8))
fig.subplots_adjust(bottom=0.15, left =0.16, right = 0.95, top = 0.95)
ax=fig.add_subplot(111)

ax.plot(r,vel,'-k',label=r'Velocidad $[v]$')
ax.plot(r,rho,'-r',label=r'Densidad $[\rho]$')
ax.plot(r,P,'-b',label=r'Presion $[P]$')
ax.plot(r,Q,'-g',label=r'Viscosidad artificial $[Q]$')

ax.minorticks_on()
ax.tick_params( labelsize=24)
ax.tick_params('both', length=3, width=1.2,which='minor', direction='in', right='on',top='on')
ax.tick_params('both', length=6, width=1.2,which='major', direction='in', right='on',top='on') 
# ax.set_ylabel(r'$\mathrm{Energy\:[M_\odot (km/s)^2}]$', fontsize=26)
ax.set_xlabel(r'$\mathrm{r/r_{tot}}$',fontsize=26)
ax.legend(fontsize=15)
plt.show()
```

# Con distintos valores de q

```python
data_20  = np.loadtxt(path+'lh2.out')
data_15  = np.loadtxt(path+'lh2_15.out')
data_10  = np.loadtxt(path+'lh2_1.out')
data_05  = np.loadtxt(path+'lh2_05.out')
data_03  = np.loadtxt(path+'lh2_03.out')
data_02  = np.loadtxt(path+'lh2_02.out')
data_01  = np.loadtxt(path+'lh2_01.out')
data_005 = np.loadtxt(path+'lh2_005.out')
data_000 = np.loadtxt(path+'lh2_000.out')
```

```python
fig=plt.figure(1, figsize=(8,8))
fig.subplots_adjust(bottom=0.15, left =0.16, right = 0.95, top = 0.95)
ax=fig.add_subplot(111)

ax.plot(data_20[:,1],data_20[:,2],'-k',label='q=2')
ax.plot(data_15[:,1],data_15[:,2],'-r',label='q=1.5')
ax.plot(data_10[:,1],data_10[:,2],'-b',label='q=1')
ax.plot(data_05[:,1],data_05[:,2],'-g',label='q=0.5')
ax.plot(data_03[:,1],data_03[:,2],'-c',label='q=0.3')
ax.plot(data_02[:,1],data_02[:,2],'-',color='darkgray',label='q=0.2')
ax.plot(data_01[:,1],data_01[:,2],'-',color='darkorange',label='q=0.1')
ax.plot(data_005[:,1],data_005[:,2],'-',color='magenta',label='q=0.05')
ax.plot(data_000[:,1],data_000[:,2],'-',color='purple',label='q=0')

ax.minorticks_on()
ax.tick_params( labelsize=24)
ax.tick_params('both', length=3, width=1.2,which='minor', direction='in', right='on',top='on')
ax.tick_params('both', length=6, width=1.2,which='major', direction='in', right='on',top='on') 
ax.set_ylabel(r'$\mathrm{E}$', fontsize=26)
ax.set_xlabel(r'$\mathrm{t}$',fontsize=26)
ax.legend(loc=4,fontsize=15)

plt.show()
```
