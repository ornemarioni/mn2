# Distribucion homogena de particulas en r

```python
n = 1000
p = np.ndarray([n,4])

for i in range(n):
    
    r3 = rm.uniform(0, 1./3.)
    ct = rm.uniform(-1, 1)
    fi = rm.uniform(0, 2*np.pi)
    
    st = np.sqrt(1 - ct**2)
        
    x = (r3*3)**(1/3.) * st * np.cos(fi)
    y = (r3*3)**(1/3.) * st * np.sin(fi)
    z = (r3*3)**(1/3.) * ct
    
    p[i,:] = [i,x,y,z]
    
#print p
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), tight_layout=True)

ax1.plot(p[:,1],p[:,2],'o')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(p[:,1],p[:,3],'o')
ax2.set_xlabel('x')
ax2.set_ylabel('z')

plt.show()
```

```python
# Calculamos la cantidad de partculas que hay entre r y r+dr. Para ello veamos en que cascara 
# se encuentra cada partcula. Suponemos que cada partcula tiene una masa m=1

dr = 0.1                  # ancho de cada cascarn
l  = int(1/dr)            # cantidad de cscaras

m_r = np.zeros(l)         # masa en cada cascarn
rm  = np.zeros(n)         # nro del cascarn al que pertenece la partcula

for i in np.arange(n):
    rm[i] = int(r[i]/dr)
    j = int(rm[i])
    m_r[j] = m_r[j] + 1   # ac sumamos las partculas que estn en cada cscara.

#print rm
#print m_r
```
